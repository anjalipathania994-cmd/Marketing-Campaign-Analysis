```python
#Data Cleaning



#Remove Extra Spaces
df = df.apply(
    lambda x: x.str.strip() if x.dtype == "object" else x
)
```

```python
df = df.drop_duplicates()

print("Duplicates removed successfully")
```

```python
df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")

print(df.columns)
```

```python
# Fill missing values with mode

df['campaign_name'] = df['campaign_name'].fillna(df['campaign_name'].mode()[0])
df['channel'] = df['channel'].fillna(df['channel'].mode()[0])
df['status'] = df['status'].fillna(df['status'].mode()[0])
```

```python
# Fill numerical missing values with median

df['leads_generated'] = df['leads_generated'].fillna(df['leads_generated'].median())
df['conversions'] = df['conversions'].fillna(df['conversions'].median())
```

```python
# Convert region names into proper format

df['region'] = df['region'].str.strip().str.title()

print(df['region'].unique())
```

```python
df.columns
```

```python
# Convert status into proper format

df['status'] = df['status'].str.strip().str.capitalize()

print(df['status'].unique())
```

```python
# Clean column names
df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")

print(df.columns)
```

```python
# Convert channel column into string
df['channel'] = df['channel'].astype(str)

# Remove extra spaces
df['channel'] = df['channel'].str.strip()

# Convert into proper format
df['channel'] = df['channel'].str.title()

# Replace invalid values with NaN
df['channel'] = df['channel'].replace(['Nan', 'Unknown', ''], np.nan)

# Fill missing values using mode
df['channel'] = df['channel'].fillna(df['channel'].mode()[0])

# Check final output
print(df['channel'].unique())
```

```python
# Convert campaign_name column into string
df['campaign_name'] = df['campaign_name'].astype(str)

# Remove extra spaces
df['campaign_name'] = df['campaign_name'].str.strip()

# Convert into proper format
df['campaign_name'] = df['campaign_name'].str.title()

# Replace invalid values with NaN
df['campaign_name'] = df['campaign_name'].replace(
    ['Nan', 'Unknown', ''], 
    np.nan
)

# Fill missing values using mode
df['campaign_name'] = df['campaign_name'].fillna(
    df['campaign_name'].mode()[0]
)

# Check final output
print(df['campaign_name'].unique())
```

```python
df['start_date'] = pd.to_datetime(
    df['start_date'],
    format='mixed',
    errors='coerce'
)
```

```python
# Convert region column into string
df['region'] = df['region'].astype(str)

# Remove extra spaces
df['region'] = df['region'].str.strip()

# Convert into proper format
df['region'] = df['region'].str.title()

# Replace invalid values with NaN
df['region'] = df['region'].replace(
    ['Nan', 'Unknown', ''],
    np.nan
)

# Fill missing values using mode
df['region'] = df['region'].fillna(
    df['region'].mode()[0]
)

# Check final output
print(df['region'].unique())
```

```python
# Convert budget column into string
df['budget_usd'] = df['budget_usd'].astype(str)

# Remove $, USD and commas
df['budget_usd'] = df['budget_usd'].str.replace('$', '', regex=False)
df['budget_usd'] = df['budget_usd'].str.replace('USD', '', regex=False)
df['budget_usd'] = df['budget_usd'].str.replace(',', '', regex=False)

# Remove extra spaces
df['budget_usd'] = df['budget_usd'].str.strip()

# Replace invalid values with NaN
df['budget_usd'] = df['budget_usd'].replace(
    ['TBD', 'Nan', ''],
    np.nan
)

# Convert into numeric datatype
df['budget_usd'] = pd.to_numeric(
    df['budget_usd'],
    errors='coerce'
)

# Fill missing values using median
df['budget_usd'] = df['budget_usd'].fillna(
    df['budget_usd'].median()
)

# Check output
print(df['budget_usd'].head())
print(df['budget_usd'].dtype)
```

```python
# Convert roi_percent column into string
df['roi_percent'] = df['roi_percent'].astype(str)

# Remove % sign
df['roi_percent'] = df['roi_percent'].str.replace(
    '%',
    '',
    regex=False
)

# Remove extra spaces
df['roi_percent'] = df['roi_percent'].str.strip()

# Replace invalid values with NaN
df['roi_percent'] = df['roi_percent'].replace(
    ['Nan', 'Unknown', ''],
    np.nan
)

# Convert into numeric datatype
df['roi_percent'] = pd.to_numeric(
    df['roi_percent'],
    errors='coerce'
)

# Fill missing values using median
df['roi_percent'] = df['roi_percent'].fillna(
    df['roi_percent'].median()
)

# Check output
print(df['roi_percent'].head())
print(df['roi_percent'].dtype)
```

```python
# Find missing campaign IDs
missing_mask = df['campaign_id'].isnull()

# Generate new unique IDs
df.loc[missing_mask, 'campaign_id'] = [
    f"CMP-{i:05d}" 
    for i in range(10001, 10001 + missing_mask.sum())
]

# Check again
print(df['campaign_id'].isnull().sum())

# Preview
print(df['campaign_id'].head(20))
```

```python
num_cols = ['budget_usd', 'leads_generated', 'conversions', 'roi_percent']

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    before = len(df)
    df = df[(df[col] >= lower) & (df[col] <= upper)]
    print(f"{col}: removed {before - len(df)} outlier rows")

```

```python
columns = ['budget_usd', 'leads_generated', 'conversions', 'roi_percent']

for col in columns:
    
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df = df[
        (df[col] >= lower) &
        (df[col] <= upper)
     ]

print(df.shape)
```
